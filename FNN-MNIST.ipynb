```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the dataset
iris = load_iris()

# Create a DataFrame for easier exploration
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Display the first 5 rows
print(iris_df.head())
```

```python
# Features are the measurements
X = iris.data

# Labels are the flower species (0, 1, or 2)
y = iris.target
```

```python
from tensorflow.keras.utils import to_categorical

# y is currently [0, 1, 2, ...]. We need to convert it.
# e.g., 0 -> [1, 0, 0], 1 -> [0, 1, 0], 2 -> [0, 0, 1]
y_categorical = to_categorical(y) #converts integer labels into one-hot encoded vectors

print("Original y shape:", y.shape)
print("One-hot encoded y shape:", y_categorical.shape)
print("\nFirst 5 original labels:\n", y[:5])
print("\nFirst 5 one-hot encoded labels:\n", y_categorical[:5])
```

```python
from sklearn.model_selection import train_test_split

# Split data into 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y_categorical, test_size=0.2, random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Assume X_train is available from the previous session
# The number of features determines the input shape
n_features = X_train.shape[1] # This will be 4 for the Iris dataset

# Define the model
model = keras.Sequential(
    [
        # Input Layer: We specify the input shape based on our data
        layers.Dense(8, activation="relu", input_shape=(n_features,)),
        
        # Hidden Layer
        layers.Dense(8, activation="relu"),
        
        # Output Layer: 3 units for 3 classes, softmax for probability distribution
        layers.Dense(3, activation="softmax"),
    ]
)

# It's always a good practice to print the summary
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',   # or the correct loss for your problem
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=1,
    validation_data=(X_test, y_test),
    verbose=0
)
```

```python
# Continue in the same notebook from the previous sessions
# Ensure X_train, y_train, X_test, y_test, and the compiled model are available

# Train the model
# We store the training history to analyze it later
history = model.fit(
    X_train, 
    y_train, 
    epochs=50, 
    batch_size=1,
    validation_data=(X_test, y_test),
    verbose=0 # Set to 0 to hide epoch logs for cleaner output in labs
)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert the history object to a pandas DataFrame for easier plotting
history_df = pd.DataFrame(history.history)

# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_df['accuracy'], label='Train Accuracy')
plt.plot(history_df['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_df['loss'], label='Train Loss')
plt.plot(history_df['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assume model, X_test, y_test, and iris (from load_iris) are available
# Get raw probability predictions
predictions_raw = model.predict(X_test)
# Convert probabilities to class indices (0, 1, or 2)
predicted_classes = np.argmax(predictions_raw, axis=1)
true_classes = np.argmax(y_test, axis=1)

# 1. Print Classification Report
print("--- Classification Report ---\n")
print(classification_report(true_classes, predicted_classes, target_names=iris.target_names))

# 2. Generate and Plot Confusion Matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.title('Confusion Matrix', fontsize=14)
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_sample_image

# Load a sample image and convert to float32
china = load_sample_image("china.jpg").astype(np.float32) / 255.0

# Convert to grayscale
image = tf.image.rgb_to_grayscale(china)

# Add batch dimension
image = tf.expand_dims(image, axis=0)

# Edge detection kernel
vertical_edge_kernel = tf.constant([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
], dtype=tf.float32)

# Reshape kernel
kernel = tf.reshape(vertical_edge_kernel, [3, 3, 1, 1])

# Convolution
feature_map = tf.nn.conv2d(
    image,
    kernel,
    strides=[1, 1, 1, 1],
    padding="SAME"
)

# Display original image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image[0, :, :, 0], cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Display feature map
plt.subplot(1, 2, 2)
plt.imshow(feature_map[0, :, :, 0], cmap="gray")
plt.title("Feature Map (Vertical Edges)")
plt.axis("off")

plt.show()
```
